Trabalho sobre grafos!

```python
import networkx as nx
import matplotlib.pyplot as plt
import nxviz
```

```python
#ler o arquivo .grapml, usando método pronto da biblioteca networkx

G = nx.read_graphml('air_traffic.graphml')
```

```python
#Imprime os nós de toda a rede para que possamos ver o formato

G.nodes(data=True)
```

Nosso estudo se dará somente sobre as regiões do Brasil. Por isso precisamos excluir os nós referentes a voôs internacionais.

```python
# Identifica os nós que possuem não possuem o valor "BRASIL" no atributo "country"
nos_para_remover = [n for n, attrs in G.nodes(data=True) if attrs.get('country') != 'BRASIL']

# Remover os nós do grafo
G.remove_nodes_from(nos_para_remover)

# Salvar o grafo modificado em um novo arquivo .graphml
nx.write_graphml(G, 'grafo.graphml')
```

```python
#Imprimindo para analisar o novo formato

G.nodes(data=True)

```
